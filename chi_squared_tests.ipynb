```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from numpy.random import randint
```

Imagine you're given an unknown die and you're trying to see if the die is loaded based on a certain number of rolls (i.e. 1000 rolls), recording the number of 1's, 2's...6's that you get. How do we test with confidence whether or not the die is unfair? In other words, the distribution of the values are unequal and $$P(X) \neq \frac{1}{6}, \text{where } X=1,2,3...6$$

```python
fair_rolls = np.random.randint(1, 7, size=1000) 
```

```python
plt.hist(fair_rolls, [1,2,3,4,5,6,7], align='left')
```
